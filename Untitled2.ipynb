```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\himan\\OneDrive\\Desktop\\Csv model\\NEET_2024_RESULTS.csv")
df.head(10)
```

```python

# Descriptive Statistics
print("Descriptive Statistics:")
print(df['marks'].describe())

# Distribution of Marks
plt.figure(figsize=(10, 6))
sns.histplot(df['marks'], kde=True, bins=10)
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()

# Boxplot for Marks
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['marks'])
plt.title('Boxplot of Marks')
plt.xlabel('Marks')
plt.show()

# Center-wise Analysis (if applicable)
center_analysis = df.groupby('state')['marks'].agg(['mean', 'median', 'std'])
print("\nCenter-wise Analysis:")
print(center_analysis)

# Visualization for Center-wise Analysis
plt.figure(figsize=(10, 6))
sns.barplot(x=center_analysis.index, y=center_analysis['mean'])
plt.title('Average Marks by state')
plt.xlabel('Center Name')
plt.ylabel('Average Marks')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sort by marks in descending order
top_students = df.sort_values(by='marks', ascending=False)

# Display top students
print("Top Students based on Marks:")
print(top_students)
```

```python
top_3_students = top_students.head(10)
print("Top 3 Students based on Marks:")
print(top_3_students)
```

```python
# Filter for Rajasthan
rajasthan_students = df[df['state'] == 'RAJASTHAN']

# Sort by marks in descending order
top_rajasthan_students = rajasthan_students.sort_values(by='marks', ascending=False)

# Display top students from Rajasthan
print("Top Students from Rajasthan based on Marks:")
print(top_rajasthan_students)
```

```python
# Filter for students from Rajasthan with exactly 720 marks
rajasthan_topper_students = df[(df['state'] == 'RAJASTHAN') & (df['marks'] == 720)]

# Display the topper students from Rajasthan
print("Topper Students from Rajasthan with 720 Marks:")
print(rajasthan_topper_students)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='dummy_srlno', y='marks', data=rajasthan_topper_students, palette='viridis')

# Customize the plot
plt.title('Topper Students from Rajasthan with 720 Marks')
plt.xlabel('Student Serial Number')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.ylim(0, 750)  # Set y-axis limit to show marks clearly

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group by city and count the number of students with 720 marks
city_wise_counts = rajasthan_topper_students['city'].value_counts().reset_index()
city_wise_counts.columns = ['city', 'count']

# Create a bar plot for city-wise counts
plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='count', data=city_wise_counts, palette='viridis')

# Customize the plot
plt.title('Number of Topper Students with 720 Marks by City in Rajasthan')
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.ylim(0, max(city_wise_counts['count']) + 1)  # Adjust y-axis limit to show data clearly

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter for students in Jaipur with marks above 700
jaipur_topper_students = df[(df['city'] == 'JAIPUR') & (df['marks'] > 700)]

# Count the number of such students
num_jaipur_toppers = len(jaipur_topper_students)

# Display the result
print(f"Number of topper students in Jaipur with marks above 700: {num_jaipur_toppers}")
```

```python
# Create a bar plot to visualize the number of topper students
plt.figure(figsize=(8, 6))
sns.barplot(x=['JAIPUR'], y=[num_jaipur_toppers], palette='viridis')

# Customize the plot
plt.title('Number of Topper Students in Jaipur with Marks Above 700')
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.ylim(0, max(num_jaipur_toppers + 1, 10))  # Adjust y-axis limit to show data clearly

# Show the plot
plt.tight_layout()
plt.show()
```
